***working with only a part of a dataframe*** 



```python
#interested in the list of ages and sex of the titanic passengers
age_sex=titanic[["Age","Sex"]]
age_sex
```

```python
#interested in passengers older than 35 years
age35=titanic[titanic["Age"]>35]
age35
```

```python
titanic["Age"]>35
```

```python
#interested in passengers in cabin class 2,3
cabin23=titanic[titanic["Pclass"].isin([2,3])]
#class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]

cabin23
```

```python
#work with passenger data for which age is known
new_age=titanic[titanic["Age"].notna()]
new_age
```

***selecting specific rows and columns from a dataframe*** 

```python
#You can use a single lable, list of labels or a slice of labels
adult_names=titanic.loc[titanic["Age"]>35,"Name"]
adult_names.head()
```

```python
#I'm interested in rows 10 till 25 and columns 3-5
titanic.iloc[9:25,2:5]
```

```python
#assign the name anonymous to the the first 3 elements of the third column
titanic.iloc[0:3,3]="annonymous"
titanic.head()
```
